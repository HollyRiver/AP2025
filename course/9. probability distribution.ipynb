### 1. 베이지안에서의 P?

* 이산형 $p(x = a_i) = w_i, \sum w_i = 1$
* 연속형 $p(x) = \int_A p(x)dx \to p(3) \neq p(X = 3) = 0$

$$\begin{align}
p(x)dx = \frac{p(x ≤ X ≤ x + dx)}{dx}
\end{align}$$

> $p(x ≤ X ≤ x + dx)$와 $dx$ 모두 0으로 가지만, 분자가 클 수록 그 값이 커짐: density
