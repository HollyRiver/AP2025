# Bayesian Deep Learning: FFG

## 1. Intro

`-` Fully Factorized Gaussian

* Considered Prior

$$p(\theta) = \prod_{l=1}^L \prod_{i=1}^{D^{(l)}} \prod_{j=1}^{D^{(l-1)}}p(w_{ij}^{(l)})$$

> 모든 뉴런에 Prior를 부여. $w_{ij}$는 MLP 레이어의 가중치

`-` Posterior approximation by $q(\theta)$

## 2. Approximation and Algorithms

* Backprogation for the FFG

1. Sample mini-batch $(x, y)$ and $\epsilon$
2. Compute $w_{ij}^{(l)} = \mu_{ij}^{(l)} + \sigma_{ij}^{(l)} \epsilon_{ij}^{(l)}$
> $w$가 흔들리는 정도가 $\epsilon$에 의해서 더 커짐. 데이터가 없는 부분에서 더 커짐...
3. Evaluation loss $\it l = \frac{1}{S}\sum_{s=1}^S \log p(y~|~x, \theta^{(s)}) + \text{KL}(q(\theta)~||~p(\theta))$
> Loss function이 $\epsilon$에 의존. 확률적으로 결정
4. Backprogation
5. Updating

* Weight이 흔들리기 때문에 결과가 달라질 수 있음.
