## 중간고사 문제 풀이

* 베이지안은 알지 못하는 모든 것을 확률 변수로 취급한다.
> 모수 이외에도 관측되지 않은 변수가 존재할 수 있음: 잠재 변수 (IQ같은 것. 실제 값을 관측할 수 없음. 관측할 수 없는 잠재 변수)
>
> 데이터 observation (known)와 모수 (unknown). 베이지안은 모수가 random이기 때문에 알아내지 못한다고 가정.
>
> 관측된 것은 고정, 관측되지 않은 모든 것은 랜덤: 모수, 잠재 변수, 결측치. 각 quantity에 대해서 frequantist는 해당하는 모형들을 다 따로 만들지만, 베이지안은 그냥 한꺼번에 다 랜덤한 확률변수로 보고 밀어넣음. inference에서까지

* 베이지안 예측 $p(y'~|~y)$은 Prior에 의존한다.
> $p(y'~|~y) = \int p(y'~|~\theta) p(\theta~|~y) d\theta$. 베이지안은 사전확률과 모두 관련되어 있다.

* 확률변수의 정의 상, 상수도 확률변수로 정의할 수 있음
> $y \sim N(\theta, 0),~ \theta \sim N(0, 1)$이라면, $y$의 분산은 증가함

* $p(\theta) = h(\theta) ↔ p(k(\theta)) = h(k(\theta))$라면 Prior $p$는 척도 불변

* Transent markov chain: markov chain이 무조건 정상분포를 가지지는 않음. 당연함.

* MH algorithm에서 타겟이 되는 분포를 완벽하게 알지 못해도 됨. 커널만 알아도 됨.

* Posterior distribution

$$\begin{align}
p(\theta~|~y_1 = 1, y_2 = 2) & \propto p(y_1 = 1, y_2 = 2 ~|~\theta)p(\theta) \\
& \propto p(y_2 = 2~|~\theta)p(\theta~|~y_1 = 1)
\end{align}$$
