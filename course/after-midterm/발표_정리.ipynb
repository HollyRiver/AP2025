## 3조: 베이지안 접근법을 활용한 Domain Adaptation

* 자율주행 환경에서 명료한 환경에서 침침한 환경으로의 전이
* 소스 데이터와 타겟 데이터의 분포가 다름

`-` 알고리즘

* 새로운 알고리즘 제안: VAE 사용 + MC Dropout으로 Uncertainty 측정

> 분포의 Alignment가 최우선 목표 -> Discriminator가 분포를 잘 맞추도록

* 4개 도메인으로 나눠진 사무용품 이미지 데이터셋
* VAE Alignment 사용 시 성능 향상. MC Dropout의 경우에는 모델 성능 비슷함.
